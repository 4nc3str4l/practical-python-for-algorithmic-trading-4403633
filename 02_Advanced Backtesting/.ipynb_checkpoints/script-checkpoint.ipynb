## Multiple simulations at once

```python
class SimpleClassificationUD(Strategy):
    
    threshold_buy = 5
    threshold_sell = 5
    
    def init(self):
        self.model = model_dt
        self.already_bought = False

    def next(self):
        explanatory_today = self.data.df.iloc[-1:, :]
        forecast_tomorrow = self.model.predict(explanatory_today)[0]
        
        if forecast_tomorrow > self.threshold_buy and self.already_bought == False:
            self.buy()
            self.already_bought = True
        elif forecast_tomorrow < self.threshold_sell and self.already_bought == True:
            self.sell()
            self.already_bought = False
        else:
            pass
```

### Define initial conditions

```python
bt = Backtest(df_explanatory, SimpleClassificationUD,
              cash=10000, commission=.002, exclusive_orders=True)
```

```python
import numpy as np
```

```python
thresholds = np.arange(1,11)
```

```python
thresholds_buy = thresholds.tolist()
```

```python
thresholds_sell = (-thresholds).tolist()
```

```python
result = bt.optimize(
    threshold_buy = thresholds_buy, threshold_sell = thresholds_sell,
    return_heatmap=True, maximize='Return [%]'
)
```

```python
df_result = result[1].reset_index()
```

```python
df_result = df_result.pivot(index='threshold_buy', columns='threshold_sell', values='Return [%]')
```
